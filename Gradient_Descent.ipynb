## Single Variable optimization

##### A piece of cardboard measuring 20cm by 20cm is to be cut so that it can be folded into a box without a lid. We want to find the size of the cut $x$ that maximizes the volume of the box. So we're basially finding the value for $x$ that maximuzes the volume of a prism by cutting corners out of each of its sides equal to $2x$, and then folding each of its sides to make a 3d prism.

# **a)** Write a function for the volumns of the box resulting from a cut of size $x$. What is the domain of the function?


The formula for volume of a prism is $v(x)=lwh$, we can consider the $l \times w$ as $(20-2x) \cdot (20-2x)$, and the height as $x$.


$$v(x) = x(20-2x)^2$$


$$x(20-2x)^2 > 0$$

$$x>0$$

$$20-2x>0 \xrightarrow{} x< 10 $$

$$\{x :x \in \mathbb{R},  0<x<10\}$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,15, 1000)

def v(x):

  return (x * ((20 - 2*x) **2) )


plt.plot(x, v(x))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['v(x)'])
plt.grid()
plt.show()



# In this graph we can clearly see the convex and concave shapes of our function v.
# Where the global maximum on the closed interval from roughly 0 to 4 represents
# the maxiumum value for x that maximizes the volume of our prism.
```

We want to find where the derivative of our objective function is equal to zero. That will tell the where the maximum and minimum values for $x$ which satisfy the contraints of our optimization problem. Above this can be achieved by the extreme value theorem, but we want to use a numerical way to achieve the same result.

The value for $x$ which maximizes the volume of our cardboard box is 3.33, as evidanced by global extrema of our function $v(x)$. Thus the maximum volume is 592 units cubed. The value $x=10$ represents the value of $x$ which collapses the volume to zero, as such we're not interested in minimizing the volume so we can ingnore that value.
